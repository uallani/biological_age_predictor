```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for plots
sns.set(style='whitegrid')
```

```python
df = pd.read_pickle('../data/final_preprocessed_dataset.pkl')
```

```python
print(df.columns.tolist())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example: BMI vs Age
sns.lineplot(data=df, x='Chronological_Age', y='BMI')
plt.title('BMI vs Chronological Age')
plt.show()
```

```python
# Optional: Convert gender to readable format
df['Gender_Label'] = df['gender'].map({1: 'Male', 2: 'Female'})

# Plot: Chronological Age vs BMI grouped by gender
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Chronological_Age', y='BMI', hue='Gender_Label')
plt.title('BMI vs Chronological Age by Gender')
plt.xlabel('Chronological Age (years)')
plt.ylabel('BMI')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# Optional: Map gender for readability
df['Gender_Label'] = df['gender'].map({1: 'Male', 2: 'Female'})

# Plot: Chronological Age vs Systolic Blood Pressure
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Chronological_Age', y='ap_hi', hue='Gender_Label')
plt.title('Systolic Blood Pressure (ap_hi) vs Chronological Age by Gender')
plt.xlabel('Chronological Age (years)')
plt.ylabel('Systolic BP (ap_hi)')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python

# Map gender labels (if not already present)
df['Gender_Label'] = df['gender'].map({1: 'Male', 2: 'Female'})

# Plot: Chronological Age vs Diastolic BP
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x='Chronological_Age', y='ap_lo', hue='Gender_Label')
plt.title('Diastolic Blood Pressure (ap_lo) vs Chronological Age by Gender')
plt.xlabel('Chronological Age (years)')
plt.ylabel('Diastolic BP (ap_lo)')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

biomarker_cols = ['Chronological_Age', 'BMI', 'ap_hi', 'ap_lo', 'Pulse_Pressure', 'MAP', 'SBP_DBP_Ratio', 'cholesterol', 'gluc']

corr = df[biomarker_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
